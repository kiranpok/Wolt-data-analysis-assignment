```python
# Import pandas library
import pandas as pd

# Load the data into a pandas DataFrame
data = pd.read_csv("dataset_for_datascience_assignment.csv")

# View first 5 rows
data.head()
```

```python
# Checking missing values
data.info()

# Describe columns
data.describe()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Distribution of purchases
plt.figure(figsize=(8,6))
sns.histplot(data['PURCHASE_COUNT'], bins=50, kde=True)
plt.title('Distribution of Total Purchase Counts per User')
plt.xlabel('Number of Purchases')
plt.ylabel('Frequency')
# Saving this plot as an image file (PNG format)
plt.savefig('distribution_purchase.png', dpi=300, bbox_inches='tight')
plt.show()

# Top countries by registration
top_countries = data['REGISTRATION_COUNTRY'].value_counts().head(10)
sns.barplot(x=top_countries.index, y=top_countries.values)
plt.title('Top 10 Countries by User Registration')
plt.xlabel('Country')
plt.ylabel('Number of Users')
plt.xticks(rotation=45)
# Saving this plot as an image file (PNG format)
plt.savefig('countries_by_registration.png', dpi=300, bbox_inches='tight')
plt.show()

# Device preferences
device_usage = data['PREFERRED_DEVICE'].value_counts()
sns.barplot(y=device_usage.index, x=device_usage.values)
plt.title('Preferred Devices for Purchases')
plt.xlabel('Number of Users')
plt.ylabel('Device')
# Saving this plot as an image file (PNG format)
plt.savefig('device preferance.png', dpi=300, bbox_inches='tight')
plt.show()


# Spending patterns
sns.histplot(data['TOTAL_PURCHASES_EUR'].dropna(), bins=50, kde=True)
plt.title('Distribution of Total Spending (EUR)')
plt.xlabel('Total Spending (EUR)')
plt.ylabel('Frequency')
# Saving this plot as an image file (PNG format)
plt.savefig('spending_pattern.png', dpi=300, bbox_inches='tight')
plt.show()
```

```python
#1. User Activation Rate:
total_users = len(data)
active_users = len(data[data['PURCHASE_COUNT'] > 0])

activation_rate = (active_users / total_users) * 100
print(f'User Activation Rate: {activation_rate:.2f}%')
```

```python
#2. Customer Retention (users with more than one purchase):
retained_users = len(data[data['PURCHASE_COUNT'] > 1])

retention_rate = (retained_users / active_users) * 100
print(f'Customer Retention Rate: {retention_rate:.2f}%')
```

```python
# 3. Average Order Value:
average_order_value = data['AVG_PURCHASE_VALUE_EUR'].mean()
print(f'Average Order Value (EUR): {average_order_value:.2f}')
```

```python
#4. Device Usage Distribution:
device_distribution = data['PREFERRED_DEVICE'].value_counts(normalize=True) * 100
print(device_distribution)
```
